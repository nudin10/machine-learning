```python
import numpy as np
y = y.astype(np.uint8)
```

```python
y.info
```

```python
X_tr, X_tst, y_tr, y_tst = X[:60000], X[60000:], y[:60000], y[60000:]
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn_clf = KNeighborsClassifier()
```

```python
from sklearn.model_selection import GridSearchCV

# param_grid = [
#     {"n_estimators": [3, 10, 30], "max_features": [2, 4, 6, 8]},
#     {"bootstrap": [False], "n_estimators": [3, 10], "max_features": [2, 3, 4]}
# ]

pg = [
    {"n_neighbors": [i for i in range(1, 10)]},
    {"weights": ["uniform", "distance"]}
]
knn_cv = GridSearchCV(knn_clf, pg, cv=3)
```

```python
import numpy as np

y_tr_l = (y_tr >= 5)
y_tr_odd = (y_tr %2 ==1)
y_multi = np.c_[y_tr_l, y_tr_odd]
```

```python
knn_cv.fit(X_tr, y_tr)
```

```python
knn_cv.best_estimator_
```

```python
knn_cv.best_params_
```

```python
y_pred = knn_cv.predict(X_tst)
```

```python
from sklearn.metrics import f1_score, accuracy_score

f1 = f1_score(y_tst, y_pred, average="macro")
f1
```

```python
acc = accuracy_score(y_tst, y_pred)
acc
```
